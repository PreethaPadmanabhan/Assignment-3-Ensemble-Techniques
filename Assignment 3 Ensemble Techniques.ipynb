## Voting Classifier

```python
LR=LogisticRegression(random_state=324)
DT=DecisionTreeClassifier(random_state=42)
KNN=KNeighborsClassifier()
```

```python
classifiers=[('Logistic Regression',LR),('Decision Tree',DT),('K Nearest Neighbors',KNN)]
```

```python
for CLF_name,CLF in classifiers:
    CLF.fit(x_train,y_train)
    y_pred=CLF.predict(x_test)
    print('{:s}:{:.3f}'.format(CLF_name,f1_score(y_test,y_pred)))

    print('{:s}:{:.3f}'.format(CLF_name,recall_score(y_test,y_pred)))
```

```python
acc=accuracy_score(y_test,y_pred)
print("Accuracy",acc)
prec=precision_score(y_test,y_pred)
print("Precision Score",prec)
f1score=f1_score(y_test,y_pred)
print("F1 Score",f1score)
recall=recall_score(y_test,y_pred)
print("Recall Score",recall)
print("Confusion Matrix")
confusion_matrix(y_test,y_pred)
```
